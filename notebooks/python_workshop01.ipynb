# Analysing data from multiple files

```python
import glob
import matplotlib.pyplot as plt
```

```python
filenames = glob.glob(os.path.join("..", "data", "python-novice-inflammation-data", "data", "*1.csv"))
```

```python
for filename in filenames:
    print(filename)
    # read data
    data = np.loadtxt(fname = filename, delimiter = ",")
    # visualize
    fig = plt.figure(figsize = (10.0, 3.0))
    
    axis1 = fig.add_subplot(1,3,1) # last digit represents the index
    axis2 = fig.add_subplot(1,3,2)
    axis3 = fig.add_subplot(1,3,3)

    axis1.set_ylabel("average")
    meanval = np.mean(data, axis = 0)
    axis1.plot(meanval)

    axis2.set_ylabel("max")
    maxval = np.max(data, axis = 0)
    axis2.plot(maxval)

    axis3.set_ylabel("min")
    minval = np.min(data, axis = 0)
    axis3.plot(minval)

    # Show
    fig.tight_layout()
    plt.show()
    output_plot_path = os.path.join("..", "data", "python-novice-inflammation-data", "data")
    output_plot_name = f"{filename[-19:-3]}"
    plt.savefig(f"{output_plot_path}/{output_plot_name}pdf")
```

# Create a function

```python
def fahr_to_celc(temp):
    """ transforms a temperature in fahrenheit to celcius
    parameters:
        temp: numerical
    """
    celcius = ((temp - 32) * (5 / 9))
    return celcius
```

# Clean up visualization in function

```python
def visualize(filename):
    
    # read data
    data = np.loadtxt(fname = filename, delimiter = ",")
    # visualize
    fig = plt.figure(figsize = (10.0, 3.0))
    
    axis1 = fig.add_subplot(1,3,1) # last digit represents the index
    axis2 = fig.add_subplot(1,3,2)
    axis3 = fig.add_subplot(1,3,3)

    axis1.set_ylabel("average")
    meanval = np.mean(data, axis = 0)
    axis1.plot(meanval)

    axis2.set_ylabel("max")
    maxval = np.max(data, axis = 0)
    axis2.plot(maxval)

    axis3.set_ylabel("min")
    minval = np.min(data, axis = 0)
    axis3.plot(minval)

    # Show
    fig.tight_layout()


    plt.show()
```
